Feature Engineering

```python
categorical_val = []
continous_val = []
for column in heart_data.columns:
    print("--------------------")
    print(f"{column} : {heart_data[column].unique()}")
    if len(heart_data[column].unique()) <= 6:
        categorical_val.append(column)
    else:
        continous_val.append(column)
```

```python
categorical_val.remove('HeartDisease')
dataframes = pd.get_dummies(heart_data, columns = categorical_val)
dataframes.head(6)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
columntoscaling = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
dataframes[columntoscaling] = sc.fit_transform(dataframes[columntoscaling])
dataframes.head(6)
```

Modelling

```python
#test_train_split
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(heart_data, test_size=0.2, random_state=5)
print(train_df.shape)
print(test_df.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,recall_score, precision_score, f1_score, roc_auc_score
          
from sklearn.model_selection import cross_val_score

def print_model_scores(model, data, predictor, target):
    model.fit(data[predictor], data[target])
    predictions = model.predict(data[predictor])
    accuracy = accuracy_score(predictions, data[target])
    recall = recall_score(predictions, data[target])
    precision = precision_score(predictions, data[target])
    print('Accuracy: %s' % '{0:2}'.format(accuracy))
    print('Accuracy Percentage : %s' % '{0:2}'.format(accuracy*100))
    print('Recall: %s' % '{0:2}'.format(recall))
    print('Precision: %s' % '{0:2}'.format(precision))
    scores = cross_val_score(model, data[predictor], data[target], scoring='neg_mean_squared_error', cv=5)

    print('Cross_Validation Scores: {}'.format(np.sqrt(-scores)))
    print(f'Average RMSE: {np.sqrt(-scores).mean()}')
```

```python
target_var = ["HeartDisease"]
predictor_var = ['Age', 'Sex', 'ChestPainType', 'RestingBP', 'Cholesterol', 'FastingBS','RestingECG','MaxHR','ExerciseAngina','Oldpeak','ST_Slope']
```

```python
#Logistic_Regression

model = LogisticRegression()

print_model_scores(model, train_df, predictor_var, target_var)
```

```python
#RandomForestClassifier
import pickle

model = RandomForestClassifier(n_estimators=50)

print_model_scores(model, train_df, predictor_var, target_var)

pickle.dump(model,open('model.pkl','wb'))

```

```python
#KNearerstNeighborsClassifier

model = KNeighborsClassifier(n_neighbors = 10)

print_model_scores(model, train_df, predictor_var, target_var)
```

```python
#GaussianNaiveBayes

model = GaussianNB()

print_model_scores(model, train_df, predictor_var, target_var)
```

```python
#SupportVectorClassifier

model = SVC(C=1,random_state=None)

print_model_scores(model, train_df, predictor_var, target_var)
```
